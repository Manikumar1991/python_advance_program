```python
def meme_sum(a,b):
    a=list(str(a))
    b=list(str(b))
    max_len=max(len(a),len(b))
    a=['0']*(max_len-len(a))+a
    b=['0']*(max_len-len(b))+b
    x= int("".join((str(int(a)+int(b)) for a,b in zip(a,b))))
    return x       
```

```python
meme_sum(26, 39)
```

```python
meme_sum(122, 81)
```

```python
#Method 2
def meme_sum1(a, b):
    a, b = sorted([str(a), str(b)], key=len)
    res = ''
    for i, j in zip(a.zfill(len(b)), b):
        res += str(int(i) + int(j))
    return int(res)
```

```python
meme_sum1(122, 81)
```

```python
def prime(n):
    '''
    Driver code
    '''
    if n>1:
        for i in range(2,n):
            #print(i)
            if n%i==0:
                break
        else:
            return True
```

```python
def next_prime(n):
    #if not n%2:
     #   n += 1
    while not prime(n):
        n+= 1
    return n
```

3. If a person traveled up a hill for 18mins at 20mph and then traveled back down the same path at 60mph then their average speed traveled was 30mph.

Write a function that returns the average speed traveled given an uphill time, uphill rate and a downhill rate. Uphill time is given in minutes. Return the rate as an integer (mph). No rounding is necessary.

Examples

ave_spd(18, 20, 60) ➞ 30

ave_spd(30, 10, 30) ➞ 15

ave_spd(30, 8, 24) ➞ 12

```python
def ave_spd(t,u,v):
    t=t/60
    distance=u*t
    down_time=distance/v
    av=(2*distance)/(t+down_time)
    return av
```

```python
ave_spd(18, 20, 60) 
```

```python
ave_spd(30, 10, 30) 
```

```python
ave_spd(30, 8, 24)
```

```python
import math
def kempner(n):
    for i in range(1,n+1):
        if math.factorial(i) % n == 0:
            return i
```

```python
def boxes(lst):
    weight,box=0,1
    for i in lst:
        weight+=i
        if weight<=10:
            box=box+1
    return box
```
