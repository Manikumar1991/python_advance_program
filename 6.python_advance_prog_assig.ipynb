```python
def find_the_difference(s,t):
    for i in t:
        if i not in s:
            print(i)
```

```python
def count_datatypes(*args):
    ci=0
    cs=0
    cbol=0
    clis=0
    ct=0
    cd=0
    #res=[ci,cs,cbol,clis,ct,cd]
    for i in args:
        if type(i)==int:
            ci=ci+1
        elif type(i)==str:
            cs=cs+1
        elif type(i)==bool:
            cbol=cbol+1
        elif type(i)==list:
            clis=clis+1
        elif type(i)==tuple:
            ct=ct+1
        if type(i)==dict:
            cd=cd+1   
        res=[ci,cs,cbol,clis,ct,cd] 
    return res
```

```python
count_datatypes(1, 45, "Hi", False)
```

```python
count_datatypes([10, 20], ("t", "Ok"), 2, 3, 1)
```

```python
count_datatypes("Hello", "Bye", True, True, False, {"1": "One", "2": "Two"}, [1, 3], {"Brayan": 18}, 25, 23)
```

```python
count_datatypes(4, 21, ("ES", "EN"), ("a", "b"), False, [1, 2, 3], [4, 5, 6])
```

```python
def ones_threes_nines(n):
    temp=n
    d={}
    d["nines"]=0
    d["threes"]=0
    d["ones"]=0
    val=0
    while temp!=0:
        if 0<n<9:
            if n>=3:
                val=n//3
                d["threes"]=val
                temp=temp-(3*val)
                d["ones"]=temp
                temp=temp-temp
            elif n<3:
                d["ones"]=n
                temp=temp-temp
        if 26 >= n >= 9:
            val = n // 9
            d["nines"]=val
            temp = temp - (9 * val)
            d["ones"]=temp
            temp = temp - temp
            if temp >= 3:
                n = temp
                val = temp// 3
                d["threes"]=val
                temp = temp - (3 * val)
                d["ones"]=temp
                temp=temp-temp
    return d  
```

```python
def fib(n):
    a=0
    b=1
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        for i in range(2,n+1):
            c=a+b
            a=b
            b=c
        return c        
```
