```python
def prime(n):
    '''
    Driver code
    '''
    if n>1:
        for i in range(2,n):
            #print(i)
            if n%i==0:
                break
        else:
            return True
    #return "n is not a prime number"      
```

```python
 def filter_primes(lst):
        l=[]
        for i in lst:
            if prime(i)==True:
                l.append(i)
        return l        
```

```python
filter_primes([7, 9, 3, 9, 10, 11, 27]) 
```

```python
filter_primes([10007, 1009, 1007, 27, 147, 77, 1001, 70]) 
```

```python
def pop(lst):
    l=len(lst)
    new_lst=[]
    i=0
    for i in range(l):
        if i==0 or i<l/2:
            new_lst.append(i)
        else:
            new_lst.append(l-i-1)
        i+=1    
    return new_lst      
```

```python
pop([0, 0, 0, 0, 4, 0, 0, 0, 0]) 
```

```python
pop([0, 0, 0, 3, 0, 0, 0])
```

```python
pop([0, 0, 2, 0, 0]) 
```

```python
pop([0])
```

```python
def loves_me(x):
    l=[]
    n="Loves me"
    n1="Loves me not"
    for i in range(x):
        if i%2==0:
            l.append(n)
        else:
            l.append(n1)
    l[-1]=l[-1].upper() 
      
    return ",".join(l)       
    
```

```python
loves_me(3) 
```

```python
loves_me(6)
```

```python
loves_me(1)
```

```python
def sort_by_letter(l):
    return sorted(l, key=lambda x:str("".join([i for i in x if i.isalpha()])))
```

```python
sort_by_letter(["932c", "832u32", "2344b"])
```

```python
sort_by_letter(["99a", "78b", "c2345", "11d"])
```

```python
sort_by_letter(["572z", "5y5", "304q2"])
```

```python
def cup_swapping(lst):
    start='B'
    for i in lst:
        if start in i:
            #print("i",i)
            s=i.replace(start,"")
            #print("hi",s)
            start=s
    return start        
```

```python
cup_swapping(["AB", "CA", "AB"]) 
```
