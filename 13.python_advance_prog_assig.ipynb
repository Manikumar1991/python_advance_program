```python
def remove_letters(lst,s):
    for i in s:
        if i in lst:
            lst.remove(i)
    return lst        
```

```python
remove_letters(["s", "t", "r", "i", "n", "g", "w"], "string")
```

```python
remove_letters(["b", "b", "l", "l", "g", "n", "o", "a", "w"], "balloon") 
```

```python
remove_letters(["d", "b", "t", "e", "a", "i"], "edabit") 
```

```python
#method 2
from collections import Counter

def remove_letters(letters, word):
    remaining = (Counter(letters) - Counter(word)).keys()
    return sorted(remaining, key=letters.index)
```

```python
2. A block sequence in three dimensions. We can write a formula for this one:
Create a function that takes a number (step) as an argument and returns the amount of blocks in that step.

Examples
blocks(1) ➞ 5
blocks(5) ➞ 39
blocks(2) ➞ 12
```

```python
def blocks(step):
    if step == 0:
        return 0
    return 0.5*step**2 + 5.5*step -1

# an**2 +bn + c
# 2a = second difference
# 3a + b = second term - first term
# a + b + c = first term
# sequence = 5,12,20,29,39
# 3a + b = 7
#
# first difference = 7,8,9,10
# second diff = 1 = 2a: a = 0.5
# 3a + b = 7 : 1.5 + b = 7 ....b=5.5
# c = 5 - 5.5 -0.5 = -1
```

```python
blocks(1)
```

```python
blocks(5)
```

```python
blocks(2)
```

```python
def add_bill(s):
    s=s.split(",")
    a=0
    for i in s:
        if i.startswith('d'):
            if i.endswith('k'):
                a+=1000*int(i[1:-1])
            else:
                a=a+int(i[1:])       
    return a            
```

```python
add_bill("d20,p40,p60,d50") 
```

```python
add_bill("p30,d20,p60,d150,p360") 
```

```python
add_bill("p30,d2k,p60,d200,p360") 
```

```python
#method 2
def add_bill(money):
    return sum(int(i[1:]) for i in money.replace('k', '000').split(',') if i.startswith('d'))
```

```python
def flip_list(lst):
    return ([w[0] if type(w)==list else [w] for w in lst])
```

```python
#method 2
def flip_list(lst):
    l=[]
    for i in lst:
        if type(i)==list:
            l.append(i[0])
        else:
            l.append([i])
    return l   
```

```python
flip_list([1, 2, 3, 4]) 
```

```python
flip_list([[5], [6], [9]]) 
```
